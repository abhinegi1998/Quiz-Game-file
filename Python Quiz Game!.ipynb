```python
q1 = """Who is Prime Minister of the India ?
a. Dr. Manmohan Singh
b. Narendra Modi
c. Rahul Gandhi
d. Mamta Banergi
"""

q2 = """Where is the Taj Mahal ?
a. Agra
b. Jaipur
c. Pune
d. Hydrabad
"""

q3 = """What is the output of 'a'+'bc' ?
a. a
b. bc
c. bca
d. abc
"""

q4 = """What is the output of 3*1**3 ?
a. 27
b. 9
c. 3
d. 1
"""
q5 = """DRDO is establish on ?
a. 1958
b. 1960
c. 1948
d. 1955
"""

questions = {q1:"b",q2:"a",q3:"d",q4:"c",q5:"a"}

name = input("Enter your name: ")
score = 0

for i in questions:
    print(i)
    ans = input("Enter the answer (a/b/c/d): ")
    if ans == questions[i]:
        print("Correct answer, You got 2 point.")
        score = score + 2
    else:
        print("Wrong answer, You lost 1 point.")
        score = score - 1
            
print("Total score is",score)
```
